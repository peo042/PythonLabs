## Exercise 7.2

Evaluating the quality of a classifier is crucial for modern data science. Given the gold standard and the predicted label, complete the following activities:

- Write code to create the confusion matrix (Hint: the matrix is composed by four numbers: TP, TN, FP, and FN)
- Write code to calculate and print:
  - Accuracy
  - Precision
  - Recall
  - F-Score (F1)

```python
goldlabel = ['P','N','N','P','N','N','N','N','N','N','P','N','N','N','N','P','N','N','N','P','P','P','N','N','N','P']
predicted = ['P','N','N','N','N','N','N','N','N','N','N','N','N','N','N','N','N','P','N','N','P','P','N','N','N','P']
```

```python
# write code here
TP, TN, FN, FP = 0, 0, 0, 0
for g,p in zip(goldlabel, predicted):
    if g == 'P':
        if p == g:
            TP += 1
        else:
            FN += 1
    if g == 'N':
        if p == g:
            TN += 1
        else:
            FP += 1
```

```python
## Print the confusion matrix here
print(' CONFUSION MATRIX')
print('     P     N')
print('  P  {0}     {1}'.format(TP,FN))
print('  N  {0}     {1}'.format(FP,TN))
```

```python
# Calc and Print ACcuracy here
ACC = (TP + TN)/(TP+TN+FN+FP)
print('The accuracy of the model is {0:.2f}.'.format(ACC))
```

```python
# Calc and Print Precision here
PREC = TP/(TP+FP)
print('The precision of the model is {0:.2f}.'.format(PREC))
```

```python
# Calc and Print Recall here
RECALL = TP/(TP+FN)
print('The recall of the model is {0:.2f}.'.format(RECALL))
```

```python
# Calc and Print F-Score F1 here
F1 = (2*PREC*RECALL)/(PREC+RECALL)
print('The f-score of the model is {0:.2f}.'.format(F1))
```

## Exercise 7.3

Calculate Cohen's kappa for the following Confusion Matrices:


# 1  


```python
# Write Cohen's kappa here
# Declare the 4 variables of the confusion matrix (see slide 52, module 15) and calculate Kappa
TP, FN, FP, TN = 25,25,25,25
po = (TP+TN)/(TP+TN+FN+FP)
pe = ((TP+FP)*(TP+FN)+(TN+FN)*(TN+FP))/((TP+TN+FN+FP)**2)
kappa = (po-pe)/(1-pe)
print(kappa)
```
